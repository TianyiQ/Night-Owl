```python
import scipy.io as scio
import numpy as np
data = scio.loadmat('./old_wider_face_val.mat')
print(type(data))
print(data.keys())
fake = {'event_list':[], 'face_bbx_list':[], 'file_list':[]}
```

```python
# print(len(data['event_list']))
# print(data['event_list'])
fake['event_list'] = np.array([[np.array(['0--Whatever'])]])
print(fake['event_list'])
```

```python
# print(data['file_list'])
cont = []
with open('./wider_val.txt', 'r') as widerval:
    lines = widerval.readlines()
    for line in lines:
        name = line.split(' ')[0]
        name = name.split('/')[-1].split('.')[0]
        cont.append([np.array([name])])
fake['file_list'] = np.array([[cont]])
```

```python
# print(data['face_bbx_list'])

with open('./wider_val.txt', 'r') as oneline:
    onelines = oneline.readlines()

cont = []

for line in onelines:
    eles = line.split(' ')
    assert eles[1][0] in '0123456789'
    cur = []
    for i in range(int(eles[1])):
        a,b,c,d = int(eles[i*5+2]), int(eles[i*5+3]), int(eles[i*5+4]), int(eles[i*5+5])
        cur.append([a,b,c,d])
    cont.append([np.array(cur)])

fake['face_bbx_list'] = np.array([[cont]], dtype=object)
# print(fake['face_bbx_list'])
```

```python
scio.savemat('./wider_face_val.mat', fake)
```
